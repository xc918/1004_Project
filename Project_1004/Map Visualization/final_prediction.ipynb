```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Prediction_Final.csv')
```

```python
data.columns = ['x','LocationID_new','Prediction']
```

```python
data = data.drop('x', axis = 1)
```

```python
data.head()
```

```python
extend_1 = np.array(range(1, 195))
extend_2 = pd.DataFrame()
```

```python
extend_2['LocationID_new'] = extend_1
```

```python
result = pd.merge(extend_2, data, on='LocationID_new', how='left')
```

```python
result = result.fillna('Low Trips')
```

```python
result.to_csv('prediction.csv')
```

```python
newdata = pd.read_csv('prediction copy.csv')
```

```python
newdata.columns = ['x','LocationID_new','Prediction']
newdata
```

```python
for i in range(len(newdata)):
    if newdata.Prediction[i] == 'Low Trips':
        newdata.x[i] =0
    elif newdata.Prediction[i] =='Taxi Dominates':
        newdata.x[i] = 1
    elif newdata.Prediction[i] == 'Uber Dominates':
        newdata.x[i] = 2
    else:
        newdata.x[i] = 3
```
